```python
df['type'].value_counts()
```

```python
import matplotlib.pyplot as plt
df['type'].value_counts().plot(kind = 'bar',color =['red','blue'])
plt.title("Count of movies vs Tv shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show
```

```python
import matplotlib.pyplot as plt
df['country'].value_counts().head(10).plot(kind = 'barh',color ='purple')
plt.title("Top 10 countries with the most content")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show
```
